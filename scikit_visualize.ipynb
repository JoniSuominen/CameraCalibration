```python
import matplotlib.pyplot as plt
import numpy as np

from sklearn.linear_model import Ridge
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import PolynomialFeatures, SplineTransformer
fig, axes = plt.subplots(ncols=1, nrows=3, figsize=(10, 10))
x_plot = np.linspace(-1, 11, 200)

x_train = np.linspace(0, 10, 200)
rng = np.random.RandomState(0)
x_train = np.sort(rng.choice(x_train, size=20, replace=False))

# create 2D-array versions of these arrays to feed to transformers
X_train = x_train[:, np.newaxis]
X_plot = x_plot[:, np.newaxis]

for i in range(0,3):
    splt = SplineTransformer(n_knots=4, degree=i+1).fit(X_train)
    axes[i].plot(x_plot, splt.transform(X_plot))
    axes[i].set_title(f"B-Spline basis of degree {i + 1}", fontweight="bold")
fig.tight_layout()

plt.savefig('splines.eps', bbox_inches='tight')
```
